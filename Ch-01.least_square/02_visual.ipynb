# Visualize 

데이터 시각화는 모델 학습에 있어 중요한 부분중 하나이다.

시각화를 통해 어떤 데이터가 잘 학습되고 있는지 눈으로 파악하기 쉽다.

(1)matplotlib를 이용하여 시각화 하기.

```python
import matplotlib.pyplot as plt
import numpy as np

import numpy as np

# 데이터 불러오기.
point = np.genfromtxt("./data/point.csv", delimiter = ',')
x = point[0]
y = point[1]
x = x.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
plt.figure(figsize = (8,8))
plt.scatter(x, y, marker = '.', color = "b")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
