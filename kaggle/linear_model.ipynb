```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data_path = './data/melb_data.csv'
data = pd.read_csv(data_path)
data
```

```python
data = data.dropna(axis = 0)
data
```

```python
y = data.Price
y = np.array(y)
y = y.reshape(-1,1)
y.shape
```

```python
x = data[['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']]
x
```

```python
x = np.array(x)
```

```python
class MyModel:
    def __init__(self, lr = 0.000001):
        self.lr = lr
        self.w = np.random.randn(5, 1)
        self.b = np.zeros((5, 1))

    def predict(self, feature):
        y_hat = feature @ self.w
        return y_hat
    
    def get_loss(self, y, y_hat):
        loss = (1 / (y.shape[0])) * ((y - y_hat).T @ (y - y_hat))
        return loss
    
    def get_gradient(self, feature, y_hat, y):
        error = y_hat - y
        grad = (1 / y.shape[0]) * (feature.T @ error)
        return grad

    
    def update_weight(self, grad):
        self.w -= self.lr * grad
loss_p = []
model = MyModel()
for i in range(500):
    y_hat = model.predict(x)
    loss = model.get_loss(y, y_hat)
    loss_p.append(loss.item())
    grad = model.get_gradient(x, y_hat, y)
    model.update_weight(grad)
    if i % 10 == 0:
        print(f"loss value : {loss}")
        print(f"predictions = {y_hat}")
    
```

```python
plt.figure(figsize = (8,8))
plt.plot(loss_p, color = 'r')

```
